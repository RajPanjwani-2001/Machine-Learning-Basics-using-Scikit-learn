```python
x = train.drop('price_range', axis=1)
y = train['price_range']
```

```python
models = [DecisionTreeClassifier(), RandomForestClassifier()]
```

```python
accuracy = []

for model in models:
    cv = cross_val_score(model, x, y, scoring='accuracy', cv=5).mean()
    accuracy.append(cv)
```
